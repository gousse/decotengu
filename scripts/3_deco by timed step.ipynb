```python
import decotengu
from decotengu import Conveyor

engine = decotengu.Engine()
engine.add_gas(0,21)

# calculate step every x min
x = 0.1
engine.calculate = Conveyor(engine, x)
```

```python
print("default model is ",engine.model)
# model :
# decotengu.ZH_L16B_GF  default is B
## engine.model=decotengu.ZH_L16B_GF()

# use C Model ZH_L16C_GF
# engine.model=decotengu.ZH_L16C_GF()

engine.model.gf_low=0.8
engine.model.gf_high=0.85
```

```python
profile = list(engine.calculate(35,40))
```

```python
print(profile)
```

```python
for step in profile:
    print(step)
```

```python
profile[0].data
```

```python
curve=list()

i=0
for step in profile:
    print(i,"time","{:5.2f}".format(step.time),"phase",step.phase, "abs_p", step.abs_p,"depth", engine._to_depth(step.abs_p))
    i+=1
    curve.append((step.time,engine._to_depth(step.abs_p)))
```

```python
for stop in engine.deco_table:
    print(stop)
```

```python
for point in curve:
    print(point)
```
