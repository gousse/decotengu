```python
import decotengu
from decotengu import Conveyor

engine = decotengu.Engine()
engine.add_gas(0,21)

# calculate step every x min
x = 0.1
engine.calculate = Conveyor(engine, x)
```

```python
print("default model is ",engine.model)
# model :
# decotengu.ZH_L16B_GF  default is B
## engine.model=decotengu.ZH_L16B_GF()

# use C Model ZH_L16C_GF
# engine.model=decotengu.ZH_L16C_GF()

engine.model.gf_low=0.8
engine.model.gf_high=0.85
```

```python
profile = list(engine.calculate(35,40))
```

```python
print(profile)
```

```python
for step in profile:
    print(step)
```

```python
profile[0].data
```

```python
def profile_curve(profile):
    curve=list()
    for step in profile:
        curve.append((step.time, -engine._to_depth(step.abs_p)))
    return curve
```

```python
for stop in engine.deco_table:
    print(stop)
```

```python
curve = profile_curve(profile)
for point in curve:
    print(point)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(curve, columns=['time','depth'])
df = df.set_index('time')
df.head()
```

```python
df.plot(kind='line',figsize=(14,8))
plt.title('Dive Profile')
plt.ylabel('Depth')
plt.xlabel('Time')
plt.show()
```
